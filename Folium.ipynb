Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python

# Create a folium map centered on a specific location
map_center = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
map_zoom = 5
site_map = folium.Map(location=map_center, zoom_start=map_zoom)

# Create a marker cluster group
marker_cluster = MarkerCluster()

# Iterate over the DataFrame rows
for index, row in spacex_df.iterrows():
    location_name = row['Launch Site']
    latitude = row['Lat']
    longitude = row['Long']
    value = row['class']
    
    # Define the marker color based on the value
    if value == 1:
        marker_color = 'green'
    elif value == 0:
        marker_color = 'red'
    else:
        marker_color = 'black'

    # Create a marker for each location with specified color
    folium.Marker(
        location=[latitude, longitude],
        icon=folium.Icon(color=marker_color),
        popup=location_name
    ).add_to(marker_cluster)
    
    site_map.add_child(marker_cluster)
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
#Calculate Distance between Coast Line Closet Point and 
# KSC LC-39A launch site

#Answer: Closest Point on Coast Line - Lat 28.56284 Long -80.56794
#Launch site is KSC LC-39A Lat 28.573255 Long -80.646895

coastline_lat=28.573255
coastline_lon=-80.79741

launch_site_lat=28.573255
launch_site_lon=-80.646895

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

print(distance_coastline,'km')
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
def place_marker_and_line(lat1, lon1, lat2, lon2, label, sitemap):
    # function to place marker on second coordinates and polyline
    # between points 
    
    #Place marker
      
    marker= folium.Marker(
        location=[lat2, lon2],
        icon=folium.Icon(color='gray',prefix='fa', icon='info-circle'),
        popup=label)

    sitemap.add_child(marker)
    
    #Place polyline
    coordinates = [[lat1, lon1],
    [lat2, lon2]]

    polyline=folium.PolyLine(locations=coordinates, weight=2,color='blue')
    sitemap.add_child(polyline)
    
    return sitemap
```

```python
#Closest City is Titusville
city_lat=28.6122
city_lon=-80.8076

distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

label = 'Closest city - Titusville: ' + str(round(distance_city)) + ' km'

place_marker_and_line(launch_site_lat, launch_site_lon, city_lat, city_lon, label, site_map)

site_map
```

```python
#Closest Railway is Florida East Coast Railway
rail_lat=28.56643
rail_lon=-80.80376

distance_rail = calculate_distance(launch_site_lat, launch_site_lon, rail_lat, rail_lon)

label = 'Closest Railway - Florida East Coast Railway: ' + str(round(distance_rail)) + ' km'

place_marker_and_line(launch_site_lat, launch_site_lon, rail_lat, rail_lon, label, site_map)

site_map
```

```python
#Closest Highway is FL 405
hwy_lat=28.5239
hwy_lon=-80.69115

distance_hwy = calculate_distance(launch_site_lat, launch_site_lon, hwy_lat, hwy_lon)

label = 'Closest Highway - FL 405: ' + str(round(distance_hwy)) + ' km'

place_marker_and_line(launch_site_lat, launch_site_lon, hwy_lat, hwy_lon, label, site_map)

site_map
```
